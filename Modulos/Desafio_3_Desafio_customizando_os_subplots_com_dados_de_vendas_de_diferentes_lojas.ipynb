```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df_vendas = pd.DataFrame(vendas_2022, index=lojas)
```

```python
fig, axs = plt.subplots(2,2, figsize=(12, 10))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Número de vendas das Lojas A, B, C e D ao longo dos meses do ano')

cores = ['b', 'g', 'r', 'purple']
ymin = 0
ymax = 400
for ax, loja, cor in zip(axs.flatten(), lojas, cores):
    ax.set_ylim(ymin, ymax)
    ax.plot(df_vendas.loc[loja], lw=3, color=cor)
    ax.set_title(f'Loja {loja}', fontsize=18, loc='left')
    ax.set_xlabel('Meses', fontsize=14)
    ax.set_ylabel('Número de Vendas', fontsize=14)
    ax.xaxis.set_tick_params(labelsize=12)
    ax.yaxis.set_tick_params(labelsize=12)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)


plt.show()
```
